```python
import numpy as np

def find_best_match(kbFeatures, mhFeatures):
    bestRangeStart = 0
    bestRangeScore = float('inf')  

    for i in range(len(kbFeatures) - len(mhFeatures) + 1):
        distance = 0
      
        for j in range(len(mhFeatures)):
            d = np.linalg.norm(kbFeatures[i + j] - mhFeatures[j])
            distance += d
        print(f"起始位置 {i} 的总距离为: {distance}")

   
        if distance < bestRangeScore:
            bestRangeScore = distance
            bestRangeStart = i

    return bestRangeStart, bestRangeScore

kbFeatures = np.random.rand(100, 128) 
mhFeatures = np.random.rand(10, 128)

best_start, best_score = find_best_match(kbFeatures, mhFeatures)
print(f"最佳匹配区间起始索引: {best_start}, 最小累积距离: {best_score}")
```

 # Create a search function that returns similar images based on the non-category based vectorisation of a dataset by using mobilenet (also feel free to look at other image classification models such as efficient net etc.)

```python
import os
import numpy as np
from tensorflow.keras.applications import MobileNet
from tensorflow.keras.applications.mobilenet import preprocess_input
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt


mobilenet = MobileNet(weights='imagenet', include_top=False, pooling='avg')


def vectorize_image(image_path, model):
    img = load_img(image_path, target_size=(224, 224)) 
    img_array = img_to_array(img)  
    img_array = np.expand_dims(img_array, axis=0)  
    img_array = preprocess_input(img_array)  
    feature_vector = model.predict(img_array)  
    return feature_vector.flatten() 


def vectorize_dataset(dataset_path, model):
    image_vectors = {}
    for root, _, files in os.walk(dataset_path):
        for file in files:
            if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                image_path = os.path.join(root, file)
                image_vectors[image_path] = vectorize_image(image_path, model)
    return image_vectors


def search_similar_images(query_image_path, dataset_vectors, model, top_k=5):
    query_vector = vectorize_image(query_image_path, model) 
    similarities = []
    for image_path, vector in dataset_vectors.items():
        sim = cosine_similarity([query_vector], [vector])[0][0]  
        similarities.append((image_path, sim))
    
    similarities = sorted(similarities, key=lambda x: x[1], reverse=True)
    return similarities[:top_k]


def show_similar_images(query_image_path, similar_images):
    plt.figure(figsize=(15, 5))
    plt.subplot(1, len(similar_images) + 1, 1)
    plt.imshow(load_img(query_image_path))
    plt.title("Query Image")
    plt.axis("off")
    
    for i, (image_path, sim) in enumerate(similar_images):
        plt.subplot(1, len(similar_images) + 1, i + 2)
        plt.imshow(load_img(image_path))
        plt.title(f"Sim: {sim:.2f}")
        plt.axis("off")
    plt.show()


dataset_path = 'IMAGE/Image_Animal'  
query_image_path = 'IMAGE/fox1.jpg'  


print("开始矢量化数据集...")
dataset_vectors = vectorize_dataset(dataset_path, mobilenet)

print("搜索相似图像...")
similar_images = search_similar_images(query_image_path, dataset_vectors, mobilenet, top_k=5)

print("相似图像：")
for img_path, similarity in similar_images:
    print(f"{img_path}: {similarity:.2f}")


show_similar_images(query_image_path, similar_images)
```

 # Create the search function on film based on the above search function (make it work with at least 10 different film files)

```python
import numpy as np
from moviepy import VideoFileClip
from sklearn.metrics.pairwise import cosine_similarity

def extract_features(video_path, duration, start_time=0, fps=16):
    video = VideoFileClip(video_path).subclipped(start_time, start_time + duration)
    frame_interval = 1 / fps  
    features = []
    for t in np.arange(0, duration, frame_interval):
        frame = video.get_frame(t)
        feature = frame.mean(axis=(0, 1)) 
        features.append(feature)
    return np.array(features)

# 寻找最相似片段函数
def find_most_similar_segment(target_features, source_features, segment_length):
    best_start = 0
    best_score = float('inf')  
    for i in range(len(source_features) - segment_length + 1):
        segment = source_features[i:i + segment_length]
        distance = np.linalg.norm(segment - target_features)
        if distance < best_score:
            best_score = distance
            best_start = i
    return best_start, best_score


def main():
    
    video_a_path = "../Input_Film/landscape.mp4"  # 寻找视频，替换为实际路径!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    video_b_path = "../Input_Film/LOVE.mp4"  # 被寻找视频，替换为实际路径!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!

    # 视频A的第24秒开始的3秒特征提取!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    duration_a = 3
    start_time_a = 24
    fps = 16
    features_a = extract_features(video_a_path, duration_a, start_time=start_time_a, fps=fps)

    
    video_b = VideoFileClip(video_b_path)
    duration_b = video_b.duration
    features_b = extract_features(video_b_path, duration_b, fps=fps)

    
    segment_length = features_a.shape[0]  
    best_start, best_score = find_most_similar_segment(features_a, features_b, segment_length)

    
    start_time = best_start / fps
    end_time = (best_start + segment_length) / fps

    print(f"最佳匹配片段时间段：{start_time:.2f}秒 - {end_time:.2f}秒，相似度分数：{best_score:.2f}")

   
    matched_clip = video_b.subclipped(start_time, end_time)
    matched_clip.write_videofile("../Output_Film/matched_segment1.mp4", audio_codec='aac') #！！！！！！！！！！！！！！！！！！！！！！！！！！！！！！！！

if __name__ == "__main__":
    main()
```

 # Start experimenting with searches across different data types, such as: text-image, image-film, film-text and so on. No need to do many different ones but choose a type of transition that will be usefull for your design work

```python
import os
import numpy as np
from moviepy import VideoFileClip
from sklearn.metrics.pairwise import cosine_similarity
import json
import subprocess

# 使用 main.py 处理视频并生成 JSON 文件
def process_video_with_main(video_path, output_json, model_path):
    input_file = "input_temp.txt"
    with open(input_file, "w") as f:
        f.write(os.path.basename(video_path) + "\n")

    video_root = os.path.dirname(video_path)

    command = [
        "python", "main.py",
        "--input", input_file,
        "--video_root", video_root,
        "--output", output_json,
        "--model", model_path,
        "--mode", "feature",
        "--no_cuda"
    ]

    subprocess.run(command, check=True)
    os.remove(input_file)

# 矢量化 class_names_list
def vectorize_class_names(class_names_path):
    class_names = []
    class_vectors = []

    with open(class_names_path, "r") as f:
        for line in f:
            word = line.strip()
            class_names.append(word)
            max_length = 20
            word_vector = np.array([ord(c) for c in word[:max_length]], dtype=float)
            padded_vector = np.pad(
                word_vector,
                (0, max(0, max_length - len(word_vector))),
                constant_values=0
            )
            class_vectors.append(padded_vector)

    return class_names, np.array(class_vectors)

# 找到与查询词最相似的词
def find_best_matching_class(query_vector, class_vectors, class_names):
    max_length = class_vectors.shape[1]
    query_vector = np.array([ord(c) for c in query_vector[:max_length]], dtype=float)
    query_vector = np.pad(
        query_vector,
        (0, max(0, max_length - len(query_vector))),
        constant_values=0
    )
    similarities = cosine_similarity(query_vector.reshape(1, -1), class_vectors)
    best_index = np.argmax(similarities)
    return class_names[best_index], best_index

def extract_all_matching_segments(json_path, best_class_name, video_root, output_folder):
    os.makedirs(output_folder, exist_ok=True)

    with open(json_path, "r") as f:
        data = json.load(f)

    found = False
    for segment in data:
        video_name = segment['video']
        for idx, clip in enumerate(segment['clips']):
            if clip.get('label') == best_class_name:
                found = True
                start_time = clip['segment'][0]
                end_time = clip['segment'][1]
                video_path = os.path.join(video_root, video_name)
                if not os.path.exists(video_path):
                    print(f"视频文件不存在: {video_path}")
                    continue

                video = VideoFileClip(video_path)
                output_path = os.path.join(output_folder, f"{video_name}_clip_{idx}.mp4")
                video.subclipped(start_time, end_time).write_videofile(output_path, audio_codec='aac')
                print(f"匹配片段已导出到: {output_path}")

    if not found:
        print(f"未找到与标签 {best_class_name} 匹配的片段。")


def main():
    video_a_path = "../Input_Film/landscape.mp4"  # 视频A路径!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    output_json_path = "../output1.json"  
    model_path = "resnet-34-kinetics-cpu.pth"  
    class_names_path = "class_names_list"  
    query_word = "example_query"  # 查询词!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    output_video_path = "../Output_Film/BestMatchVedio.mp4"  # 导出路径!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!

   
    process_video_with_main(video_a_path, output_json_path, model_path)

   
    class_names, class_vectors = vectorize_class_names(class_names_path)

  
    best_class_name, _ = find_best_matching_class(query_word, class_vectors, class_names)
    print(f"最相似的 class name: {best_class_name}")

   
    extract_all_matching_segments(output_json_path, best_class_name, os.path.dirname(video_a_path), output_video_path)

if __name__ == "__main__":
    main()
```
